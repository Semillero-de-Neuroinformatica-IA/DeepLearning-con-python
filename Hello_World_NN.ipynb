## Definición de la red neuronal
Las redes neuronales tienen una compposición básica, la cuál se muestra a continuación
neural_network.png

Se componen de

1.   Capa de entrada: Esta es la capa en la que se alimenta la red, y su tamaño esta relacionado con los datos de entrada.
2.   Capas ocultas: Son las capas de redes neuronales en las cuales se realiza la mayor parte del procesamiento de los datos. Sus conexiones, su número y la cantidad de neuronas que tiene una capa, son parámetros que ajusta el diseñador.
3.   Capa de salida: Esta capa es la que devuelve los datos de la red, su número depende del tipo de operación que se quiera, para el caso de la regresión suele ser solo una neurona, para el caso de la clasificación suele haber una neurona por cada clase.
